### 18. Calcula este promedio con Numpy y después con un método de Pandas

```python
# Media con numpy

media_numpy = np.mean(df['Rank'])
print(f"Media con NumPy: {media_numpy}")

# Media ignorando valores NaN (si los hubiera)
media_numpy_sin_nan = np.nanmean(df['Rank'])
print(f"Media ignorando NaN: {media_numpy_sin_nan}")
```

```python
#Media con pandas

# por defecto los métodos estadísticos como mean(), sum(), std() ya ignoran los valores NaN automáticamente
# aqui fuerzo a incluir los valores NaN con skipna=False para mostrar la diferencia
media_pandas = df['Rank'].mean(skipna=False)
print(f"Media con Pandas: {media_pandas}")

# Media ignorando valores NaN (si los hubiera)
media_pandas_sin_nan = df['Rank'].mean()
print(f"Media ignorando NaN: {media_pandas_sin_nan}")
```

### 20. Observa los ratings que te interesen y ahora filtra el dataframe con ese rating para ver cuáles son las películas con dicho rating

```python
ratings_interes = [8.5, 8.6, 8.7, 8.8, 8.9, 9.0]
df_filtrado =df['Rating'].isin(ratings_interes)
# Filtrar películas con esos ratings exactos
peliculas_filtradas = df[df_filtrado]

print(f"Películas con ratings {ratings_interes}: {len(peliculas_filtradas)}")
print(peliculas_filtradas[['Title', 'Rating', 'Year', 'Genre']].head())
```

### 21. Obten los valores únicos de la variable Rating y la frecuencia total de cada uno de estos valores. Posteriormente crea un nuevo dataframe con essos valores

```python

# Calcular frecuencia de cada valor único de rating
# valores únicos ordenados
frecuencia = df['Rating'].value_counts()
frec = df['Rating'].value_counts()
print(f"Total de valores únicos: {len(frec)}\n")
distribuccion =pd.DataFrame({
        'Rating': frec.index,
        'Frecuencia': frec.values
    })
distribuccion.head()
```

### 22. Ordena el nuevo dataframe por la variable `rating`

```python
df_ordenado=distribuccion.sort_values(by='Rating')
df_ordenado.head()
```

### 23. Crea una gráfica de barras con este nuevo dataframe ordenado

```python
plt.figure(figsize=(15, 7))

bars = plt.bar(range(len(df_ordenado)), df_ordenado['Rating'], label='Ratings',
               color='skyblue', edgecolor='black', alpha=0.7)

plt.title('Ratings de Películas (Ordenadas de Menor a Mayor)', 
          fontsize=16, fontweight='bold')
plt.xlabel('Películas (ordenadas por rating)', fontsize=12)
plt.ylabel('Rating', fontsize=12)
plt.grid(axis='y', alpha=0.3)


plt.legend()
plt.tight_layout()
plt.show()
```

### 24. Crea la matriz de correlación del dataframe de películas

```python
#solo funciona con columnas numericas
columnas_numericas = df.select_dtypes(include=[np.number]).columns
df_numerico = df[columnas_numericas]

# Calcular matriz de correlación
matriz_correlacion = df_numerico.corr()

print("=== MATRIZ DE CORRELACIÓN ===")
print(f"Dimensiones: {matriz_correlacion.shape}")
print("\nMatriz completa:")
print(matriz_correlacion.round(3))
```

### 25. Grafica la matriz de correlación utilizando `matshow()`

```python
# Gráfica básica con matshow
plt.figure(figsize=(10, 8))

# Usar matshow para mostrar la matriz
plt.matshow(matriz_correlacion, fignum=False, cmap='coolwarm')

# Personalizar
plt.title('Matriz de Correlación', fontsize=14, pad=20)
plt.colorbar(label='Coeficiente de Correlación')

# Añadir etiquetas
plt.xticks(range(len(matriz_correlacion.columns)), matriz_correlacion.columns, rotation=45, ha='right')
plt.yticks(range(len(matriz_correlacion.columns)), matriz_correlacion.columns)

plt.tight_layout()
plt.show()
```
